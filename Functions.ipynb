**Q3. What are the different ways to define and call a function in Python?**
   -->There are several ways to define and call a function:
   1. Standard Function
   2. Function with Parameters
   3. Function with Return Value
   4. Default Parameter Function
   5. Lambda Function
   6. Function with Function Variable Arguments

```python
#1.Standard Function – Defined using def keyword and called by its name.
def greet():
    print("Radhe Radhe!")
greet()  # Calling

#2.Function with Parameters – Accepts inputs when called.
def greet(name):
    print(f"Radhe Radhe, {name}!")
greet("Shagun")  # Calling with argument

#3.Function with Return Value – Sends a value back to the caller.
def add(a, b):
    return a + b
result = add(3, 5)  # Calling and storing result

#4.Default Parameter Function – Uses default values if no argument is passed.
def greet(name="Madhav"):
    print(f"Radhe Radhe, {name}!")
greet()         # Uses default
greet("Madhav")   # Overrides default

#5.Lambda Function – A small anonymous function using lambda keyword.
square = lambda x: x * x
print(square(8))  # Calling

#6.Function with Variable Arguments – Accepts multiple arguments using *args or **kwargs.
def show_numbers(*args):
    print(args)
show_numbers(1, 2, 3)  # Calling with multiple arguments
```

**Q6. Explain the concept of generators in Python and how they are defined.**
  --> A generator is a special type of function that produces values one at a time using the yield keyword instead of returning them all at once.
  Generators are used for lazy evaluation, meaning they generate values only when needed, which saves memory and improves performance for large datasets.
  #How they are defined:
  #Defined like a normal function but use yield instead of return.
  #Each call to yield pauses the function’s state, allowing it to resume from the same point later.
  #Generators automatically implement the iterator protocol (__iter__() and __next__()).

```python
#Example:

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 2

# Using the generator
for num in count_up_to(10):
    print(num)
```

**Q7. What are the advantages of using generators over regular functions?**
  --> Advantages of using Generators over Regular Functions in Python:

```python
'''
  1.Memory Efficiency :– Generators produce values one at a time instead of storing the entire result in memory.
  Example: Generating numbers up to 1 billion without storing them all at once.

  2.Lazy Evaluation :– Values are computed only when needed, which saves processing time for unused results.
  Example: Reading large files line-by-line instead of loading the whole file.

  3.Infinite Sequences Support :– Can generate endless data streams without memory overflow.
  Example: count() from itertools can generate numbers forever.

  4.Improved Performance :– Reduces the overhead of returning large data structures, making loops faster.
  Example: Iterating over yield results is faster for large datasets.

  5.Cleaner Code :– Removes the need for manual iterator classes; yield automatically maintains state.
  Example: A generator can replace a complex __iter__ and __next__ implementation.'''
```

**Q9. Explain the purpose and usage of the `map()` function in Python.**
  --> The map() function is used to apply a given function to each item of an
  iterable (like a list, tuple, or set) and returns a map object (an iterator) containing the results.

  #Purpose:

  '''
  To avoid writing manual loops when applying the same operation to multiple elements.
  Makes code shorter and more readable for element-wise transformations.
  
  '''

```python
#Syntax:
#map(function, iterable)
#function → The function to apply.
# iterable → The data to process (list, tuple, etc.).
#Example:
# Function to square a number
def square(x):
    return x * x
numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))  # Output: [1, 4, 9, 16]

#Example with Lambda:
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

**Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
  --> In Python, map(), filter(), and reduce() are higher-order functions used
  for processing iterables, but they work differently. The map() function applies a given function to each element of an iterable and returns a map object (iterator) containing the transformed elements.
  For example, map(lambda x: x*2, [1,2,3]) produces [2,4,6]. The filter() function selects elements from an iterable that satisfy a given condition and returns only those matching elements; for instance, filter(lambda x: x%2==0, [1,2,3,4]) produces [2,4].
  The reduce() function, available in the functools module, applies a function cumulatively to the elements of an iterable to reduce it to a single value, such as reduce(lambda x,y: x+y, [1,2,3,4]) which results in 10. In short, map() transforms data, filter() extracts specific data, and reduce() combines data into one result.

```python
from functools import reduce

nums = [1, 2, 3, 4,5]

# map()
mapped = list(map(lambda x: x * 5, nums))       # output = [5, 10, 15, 20, 25]

# filter()
filtered = list(filter(lambda x: x % 5 == 0, nums))  # output =  [5]

# reduce()
reduced = reduce(lambda x, y: x + y, nums)      # output =  15

print(mapped, filtered, reduced)
```

**Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list**

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:  # Check if even
            total += num
    return total

# Example usage
nums = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(nums)
print("Sum of even numbers:", result)
```

**Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Print all prime numbers from 1 to 200
for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")
```

**Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1  # First two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration

# Example usage
fib = Fibonacci(10)  # First 10 terms
for num in fib:
    print(num, end=" ")
```

**Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i  # Yield power of 2

# Example
for value in powers_of_two(5):
    print(value, end=" ")
```

**Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit**

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    try:
        # Ensure the input is a number
        return (celsius * 9/5) + 32
    except TypeError:
        print(f"Invalid input: {celsius} is not a number.")
        return None

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100, -40, 15.5]

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Display the results
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```
